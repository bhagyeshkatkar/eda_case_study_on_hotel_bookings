# Hypothesis Testing

##### 1. H0: There is no difference in ADR between bookings made through Online TA and Direct channels

```python
from statsmodels.stats import weightstats as ssw
```

```python
# Group data
online_ta_adr = df[df['distribution_channel'] == 'TA/TO']
direct_adr = df[df['distribution_channel'] == 'Direct']

# Perform z-test
zscore, pvalue = ssw.ztest(direct_adr['adr'], online_ta_adr['adr'], alternative='two-sided')
print("H0: No difference in ADR between Online TA and Direct")
print(f"Z-score: {zscore:.4f}, P-value: {pvalue:.4f}")
if pvalue < 0.05:
    print("🔴 Reject H0 → ADR differs significantly")
else:
    print("🟢 Fail to Reject H0 → No significant difference")
```

- A Z-test is generally used when the population standard deviation is known, or when the sample size is large (typically n>30)

##### 2. H0: Room upgrades are independent of lead time

```python
from statsmodels.stats.weightstats import ttest_ind

# Create room_reassigned column
df['room_reassigned'] = (df['assigned_room_type'] != df['reserved_room_type']).astype(int)

# Group lead_time values
lead_time_reassigned = df[df['room_reassigned'] == 1]['lead_time']
lead_time_not_reassigned = df[df['room_reassigned'] == 0]['lead_time']

# Perform Welch’s t-test
t_stat, p_val, _ = ttest_ind(lead_time_reassigned, lead_time_not_reassigned, usevar='unequal')

# Print result
print("\nH0: Room upgrades are independent of lead time")
print(f"T-statistic: {t_stat:.4f}, P-value: {p_val:.4f}")
if p_val < 0.05:
    print("🔴 Reject H0 → Lead time influences room upgrades")
else:
    print("🟢 Fail to Reject H0 → No significant relationship")
```

- Based on Welch’s t-test (p < 0.05), we reject the null hypothesis and conclude that lead time significantly affects room reassignment. This implies that guests who book earlier are more likely to receive upgraded or different room types, possibly due to operational adjustments or availability buffers created by early planning.

##### 3. H0: Average stay duration does not differ between customer types

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm

# Total stay
df['Total_nights_stayed'] = df['stays_in_week_nights'] + df['stays_in_weekend_nights']

# OLS model
model = ols('Total_nights_stayed ~ C(customer_type)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print("\nH0: Average stay duration does not differ between customer types")
print(anova_table)

if anova_table["PR(>F)"][0] < 0.05:
    print("🔴 Reject H0 → Stay duration differs by customer type")
else:
    print("🟢 Fail to Reject H0 → No significant difference")
```

- There is a statistically significant difference in the average total stay duration across different customer types (e.g., Contract, Group, Transient, Transient-party).

- Result: Fail to Reject H0

- Conclusion: ADR is similar across both channels.

2️⃣ Room Upgrades vs Lead Time
- Test Used: Welch’s t-test

- H0: Room upgrades are independent of lead time

- Result: Reject H0

- Conclusion: Early bookings are more likely to get upgrades.

3️⃣ Stay Duration vs Customer Type
- Test Used: One-way ANOVA

- H0: Stay duration does not differ by customer type

- Result: Reject H0
